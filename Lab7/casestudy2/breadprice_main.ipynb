```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('breadprice.csv')

df = df.fillna(0)  # Replace missing values with 0
df['Year'] = df['Year'].astype(int)  # Convert 'Year' column to integer

# Calculate the average price for each year
avg_prices = df.mean(axis=1)

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], avg_prices)
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Bread Price Over Time')
plt.grid(True)
plt.show()
```
